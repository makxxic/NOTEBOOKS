```python
import numpy as np
import matplotlib.pyplot as plt

def make_sample_data(n=20, seed=42, slope=3.5, intercept=2.0, noise_std=2.0):
    """
    Create a small synthetic dataset y = slope * x + intercept + noise
    Returns x (shape (n,)), y (shape (n,))
    """
    rng = np.random.default_rng(seed)
    x = np.linspace(0, 10, n)
    noise = rng.normal(loc=0.0, scale=noise_std, size=n)
    y = slope * x + intercept + noise
    return x, y
```

```python
# ==============================================
# Linear Regression with One Variable (Manual)
# ==============================================

# Example dataset
x = [1, 2, 4]
y = [2, 3, 5]

# Number of data points
n = len(x)

# ----------------------------------------------
# Step 1: Compute all required summations
# ----------------------------------------------
sum_x = sum(x)
sum_y = sum(y)
sum_x2 = sum([xi**2 for xi in x])
sum_xy = sum([x[i] * y[i] for i in range(n)])

print("Summations:")
print(f"Σx  = {sum_x}")
print(f"Σy  = {sum_y}")
print(f"Σx² = {sum_x2}")
print(f"Σxy = {sum_xy}")

# ----------------------------------------------
# Step 2: Compute slope (m) and intercept (c)
# Using the Normal Equation formulas:
# m = (nΣxy - ΣxΣy) / (nΣx² - (Σx)²)
# c = (Σy - mΣx) / n
# ----------------------------------------------
m = (n * sum_xy - sum_x * sum_y) / (n * sum_x2 - sum_x**2)
c = (sum_y - m * sum_x) / n

print("\nModel Parameters:")
print(f"Slope (m): {m:.3f}")
print(f"Intercept (c): {c:.3f}")

# ----------------------------------------------
# Step 3: Predict y values for each x
# ----------------------------------------------
y_pred = [m * xi + c for xi in x]

print("\nPredictions:")
for xi, yi, ypi in zip(x, y, y_pred):
    print(f"x={xi}, Actual y={yi}, Predicted y={ypi:.3f}")

# ----------------------------------------------
# Step 4: Calculate Mean Squared Error (MSE)
# ----------------------------------------------
mse = sum([(y[i] - y_pred[i])**2 for i in range(n)]) / n
print(f"\nMean Squared Error (MSE): {mse:.4f}")

# ----------------------------------------------
# Step 5: Plot the data and regression line
# ----------------------------------------------
import matplotlib.pyplot as plt

plt.scatter(x, y, color='red', label='Actual Data')
plt.plot(x, y_pred, color='blue', label='Fitted Line')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Manual Linear Regression')
plt.legend()
plt.grid(True)
plt.show()
```
