```python
# Import the matplotlib library for plotting
import matplotlib.pyplot as plt
```

```python
# Given a lists of X and Y values
X = [0.0, 0.52631579, 1.05263158, 1.57894737, 2.10526316,
     2.63157895, 3.15789474, 3.68421053, 4.21052632, 4.73684211,
     5.26315789, 5.78947368, 6.31578947, 6.84210526, 7.36842105,
     7.89473684, 8.42105263, 8.94736842, 9.47368421, 10.0]

Y = [2.60943416, 1.76213705, 7.18511292, 9.40744522, 5.46635068,
     8.6061673, 13.30831239, 14.26225166, 16.70323979, 16.87285951,
     22.17984858, 23.81874177, 24.23732455, 28.20185083, 28.72449237,
     27.91299402, 32.21118578, 31.39802427, 36.91479534, 36.90014818]

# 1)Define the prediction function for a linear model (y = mx + c)
def y_pred(x, m, c):
    """
    Linear regression prediction function: y = m*x + c
    This function takes a single x value, slope (m), and intercept (c) and returns the predicted y value.
    """
    # Calculate the predicted y value using the linear equation
    return m * x + c


# 2)Define the Mean Squared Error (MSE) loss function
def loss_function(y_actual, y_predicted):
    """
    Calculate the Mean Squared Error manually without numpy
    MSE is a common metric to evaluate the performance of a regression model.
    It measures the average of the squares of the errorsâ€”that is, the average squared difference between the estimated values and the actual value.
    """
    total_error = 0 # Initialize the total error to zero
    n = len(y_actual) # Get the number of data points
    # Iterate through each data point to calculate the squared error
    for i in range(n):
        # Add the squared difference between the actual and predicted y values to the total error
        total_error += (y_actual[i] - y_predicted[i]) ** 2
    # Calculate the mean squared error by dividing the total error by the number of data points
    mse = total_error / n
    return mse # Return the calculated MSE

# 3)Choose slope (m) and intercept (c) for the linear model
m = 3.0   #  this value can be modified to see how the line fits - This is the slope of the line
c = 2.0 # This is the y-intercept of the line

# Generate predicted Y values manually using the chosen m and c
Y_pred = [] # Initialize an empty list to store predicted Y values
for x in X: # Iterate through each X value
    # Calculate the predicted Y value using the y_pred function and append it to the Y_pred list
    Y_pred.append(y_pred(x, m, c))

# Calculate MSE loss using the actual and predicted Y values
mse_value = loss_function(Y, Y_pred)
# Print the calculated MSE value, formatted to 4 decimal places
print(f"Mean Squared Error (MSE): {mse_value:.4f}")

# 4)Plot scatter points of the actual data and the regression line
plt.figure(figsize=(8, 5)) # Create a new figure for the plot with a specified size
# Plot the actual data points as a scatter plot
plt.scatter(X, Y, color='blue', label='Actual Data Points')
# Plot the predicted regression line
plt.plot(X, Y_pred, color='red', label=f'Predicted Line (y = {m}x + {c})')
plt.xlabel('X values') # Set the label for the x-axis
plt.ylabel('Y values') # Set the label for the y-axis
plt.title('Manual Linear Regression') # Set the title of the plot
plt.legend() # Display the legend to identify the plotted elements
plt.show() # Show the generated plot
```
